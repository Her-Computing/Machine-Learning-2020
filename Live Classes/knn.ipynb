```python
import pandas as pd #pandas import
#sklearn imports
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split 
from sklearn import linear_model, preprocessing
import numpy as np #numpy import
```

```python
cars = pd.read_csv('car.csv') #loading in dataset
```

```python
le = preprocessing.LabelEncoder()

buying = le.fit_transform(list(cars["buying"]))
maint = le.fit_transform(list(cars["maint"]))
door = le.fit_transform(list(cars["door"]))
persons = le.fit_transform(list(cars["persons"]))
lug_boot = le.fit_transform(list(cars["lug_boot"]))
safety = le.fit_transform(list(cars["safety"]))
cls = le.fit_transform(list(cars["class"]))
```

```python
#gets the x and y coordinates 
x = list(zip(buying, maint, door, persons, lug_boot, safety))
y = list(cls)
```

```python
#split into train and test data
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.15)
```

```python
model = KNeighborsClassifier(n_neighbors = 11)
model.fit(x_train, y_train) #train our model
acc = model.score(x_test, y_test) #score our model
print(acc)
```

```python
#prediciting output from model
y_predict = model.predict(x_test)
```

```python
names = ["unacc", "acc", "good", "vgood"]

for x in range(len(y_predict)):
    print("Predicted: ", names[y_predict[x]], " , ", "Data: ", x_test[x], "Actual: ", names[y_test[x]])
```
