```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 1000)

plt.rcParams.update({'font.size' : 15})

plt.plot(x, np.sin(x), color = "black", linestyle = "dotted", label = "y = sin(x)")
plt.plot(x, np.sin(x - 1), color = "blue", linestyle = "solid", label = "y = sin(x) - 1")
plt.plot(x, np.sin(x - 1.75), color = "green", linestyle = "dashdot", label = "y = sin(x) - 1.75")
plt.plot(x, np.sin(x - 1.5), color = "red", linestyle = "dashed", label = "y = sin(x) - 1.5")

plt.axis([0, 10, -1.05, 1.05])
plt.title("Graph of sin(x)")
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend(loc = "lower left")
plt.show()
```
